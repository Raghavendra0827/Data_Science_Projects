```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
books_ratings = pd.read_csv("/content/Book1.csv", encoding="ISO-8859-1")
```

```python
books_details = pd.read_csv("/content/BX-Books.csv", encoding="ISO-8859-1")
```

```python
books_details.columns = ["ISBN","Book-Title","Book-Author","Year-Of-Publication","Publisher","Image-URL-S","Image-URL-M","Image-URL-L"]
books_details["ISBN"] = books_details["ISBN"].apply(lambda x: re.sub(r'\D', '', str(x)))

books_details.head()
```

```python
books_ratings.columns = ["User-ID","ISBN","Book-Rating"]
books_ratings["ISBN"] = books_ratings["ISBN"].apply(lambda x: re.sub(r'\D', '', str(x)))
books_ratings.head()
```

```python
import math
from surprise import Reader, Dataset, SVD
from surprise.model_selection import cross_validate
```

```python
reader=Reader()
```

```python
data=Dataset.load_from_df(books_ratings[["User-ID","ISBN","Book-Rating"]][:100000], reader)
```

```python
data
```

```python
model=SVD()
```

```python
cross_validate(model, data, measures=['RMSE','MAE'], cv=4)
```

```python
books_ratings['Estimate_Score']=books_ratings['ISBN'].apply(lambda x: model.predict(276762, x).est)
```

```python
books_ratings
```

```python
lst_movie_ID = list(books_ratings.sort_values("Estimate_Score", ascending = False)["ISBN"].unique())[:5]
```

```python
lst_movie_ID
```

```python
books_titles = list()
for val in lst_movie_ID:
  books_titles.append((books_details[books_details.ISBN == val]["Book-Title"]).to_list()[0])
```

```python
books_titles
```

```python
books_urls = list()
for val1 in lst_movie_ID:
  books_urls.append((books_details[books_details.ISBN == val1]["Image-URL-S"]).to_list()[0])
```

```python
books_urls
```

```python
from IPython.display import display, Image
import requests
from PIL import Image
from io import BytesIO

def display_image_from_link_with_title(image_link, title):
    try:
        headers = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
        }

        # Fetch the image from the link with headers
        response = requests.get(image_link, headers=headers)
        response.raise_for_status()  # Check for request errors

        # Open the image using PIL
        img = Image.open(BytesIO(response.content))

        # Display the image in the notebook with the title
        display(img)
        print(f"Title: {title}")

    except Exception as e:
        print(f"Error: {e}")

# Iterate over the zipped lists and display images with titles
for image_link, title in zip(books_urls, books_titles):
    display_image_from_link_with_title(image_link, title)
```
